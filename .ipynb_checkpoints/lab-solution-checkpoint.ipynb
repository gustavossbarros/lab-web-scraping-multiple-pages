# Lab | Web Scraping Multiple Pages

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
from random import randint
```

```python
# getting the different urls

for i in range(1,5):
    print(i)
```

```python
urls = [
    "https://bittermelodies.com/2021/02/09/1000-greatest-songs-of-all-time/",
    "https://bittermelodies.com/2021/02/11/1000-greatest-songs-of-all-time-part-2-750-501/",
    "https://bittermelodies.com/2021/02/14/1000-greatest-songs-of-all-time-part-3-500-251/",
    "https://bittermelodies.com/2021/02/14/1000-greatest-songs-of-all-time-part-4-250-1/"
]
```

```python
new_urls = []
for url in urls:
    for i in range(1,6):
        new_urls.append(url+str(i)+"/")
```

```python
new_urls
```

```python
# making requests and storing the different pages

pages = []

for item in new_urls:
    response = requests.get(item)
    print('Status code:', str(response.status_code))
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    pages.append(soup)
    
    sleep_time = randint(1,4)
    print(f'sleeping for {sleep_time} seconds')
    time.sleep(sleep_time)
```

```python
len(pages)
```

```python
# scraping the pages
```

```python
pages[0].find_all('strong')[49].text
```

```python
pages[1].find_all('strong')[0].text
```

```python
content = []

for page in pages:
    for i in range(0,50):
        content.append(page.find_all('strong')[i].text)
```

```python
content
```

```python
new_content = [item for item in content if item != '']
```

```python
new_content
```

```python
import re
```

```python
new_content[0].split('–')[1]
```

```python
artist = []
songs = []
year = []

for item in new_content:
    print(re.findall('[A-z]+', item.split('–')[1]))
    artist.append(re.findall('[A-z]+', item.split('–')[0]))
    songs.append(re.findall('[A-z]+', item.split('–')[1]))
    year.append(re.findall('[0-9]+', item.split('–')[-1]))
```

```python
songs
```

```python
year
```

```python
new_artist = []
for list_ in artist:
    new_artist.append(" ".join(list_))
```

```python
new_songs = []
for list_ in songs:
    new_songs.append(" ".join(list_))
```

```python
new_year = []
for list_ in year:
    new_year.append(list_[-1])
```

```python
# putting the info together on a dataframe
```

```python
results = pd.DataFrame([new_artist, new_songs, new_year]).T
results.columns = ['artist', 'song', 'release_year']
```
